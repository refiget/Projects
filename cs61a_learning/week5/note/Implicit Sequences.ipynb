## 4.2.1 Iterators

Python and many other programming languages provide a unified way to process elements of a container value sequentially, called an iterator that provides sequential access to values, one by one.

An iterator has 2 components:
1. a mechanism for retrieving the next element is the sequence being processed
2. a mechanism for signaling that the end of the sequence has been reached and no further elements remain.

- built-in *iter* to build an iterator
- built-in *next* to get the content of an iterator
